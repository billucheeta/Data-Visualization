```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sb
sb.set(style='darkgrid')
```

```python
tips = sns.load_dataset('tips')
sb.relplot(x="total_bill",y='tip',data=tips)
```

```python
sb.relplot(x='total_bill',y='tip',hue='smoker',data=tips)
```

```python
sb.relplot(x="total_bill", y="tip", hue="smoker", style="smoker",data=tips)
```

```python
sb.relplot(x='total_bill',y='tip',hue='smoker',style='time',data=tips)
```

```python
sb.relplot(x='total_bill',y='tip',hue='size',data=tips)
```

```python
sb.relplot(x="total_bill", y="tip", hue="size", palette="ch:r=-.5,l=.75", data=tips);
```

```python
sb.relplot(x='total_bill',y='tip',hue='smoker',size='size',data=tips)
```

```python
sb.relplot(x='total_bill',y='tip',hue='smoker',size = 'size',sizes=(15,200),data=tips)
```

```python
df = pd.DataFrame(dict(time=np.arange(500),value=np.random.randn(500).cumsum()))
g = sb.relplot(x='time',y='value',kind='line',data=df)
g.fig.autofmt_xdate()
```

```python
df = pd.DataFrame(np.random.randn(500, 2).cumsum(axis=0), columns=["x", "y"])
sb.relplot(x="x", y="y", sort=False, kind="line", data=df);
```

```python
fmri = sb.load_dataset("fmri")
sb.relplot(x="timepoint", y="signal", kind="line", data=fmri);
```

```python
sb.relplot(x="timepoint", y="signal", ci=None, kind="line", data=fmri);
```

```python
sb.relplot(x="timepoint", y="signal", kind="line", ci="sd", data=fmri);
```

```python
sb.relplot(x="timepoint", y="signal", estimator=None, kind="line", data=fmri);
```

```python
sb.relplot(x="timepoint", y="signal", hue="event", kind="line", data=fmri);
```

```python
sb.relplot(x="timepoint", y="signal", hue="region", style="event",
            kind="line", data=fmri);
```

```python
sb.relplot(x="timepoint", y="signal", hue="region", style="event",
            dashes=False, markers=True, kind="line", data=fmri);
```

```python
sb.relplot(x="timepoint", y="signal", hue="event", style="event",
            kind="line", data=fmri);
```

```python
sb.relplot(x="timepoint", y="signal", hue="region",
            units="subject", estimator=None,
            kind="line", data=fmri.query("event == 'stim'"));
```

```python
dots = sb.load_dataset("dots").query("align == 'dots'")
sb.relplot(x="time", y="firing_rate",
            hue="coherence", style="choice",
            kind="line", data=dots);
```

```python
palette = sb.cubehelix_palette(light=.8, n_colors=6)
sb.relplot(x="time", y="firing_rate",
            hue="coherence", style="choice",
            palette=palette,
            kind="line", data=dots);
```

```python
from matplotlib.colors import LogNorm
palette = sb.cubehelix_palette(light=.7, n_colors=6)
sb.relplot(x="time", y="firing_rate",
            hue="coherence", style="choice",
            hue_norm=LogNorm(),
            kind="line", data=dots);
```

```python
df = pd.DataFrame(dict(time=pd.date_range("2017-1-1", periods=500),
                       value=np.random.randn(500).cumsum()))
g = sb.relplot(x="time", y="value", kind="line", data=df)
g.fig.autofmt_xdate()
```
